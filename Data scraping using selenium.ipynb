```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import pandas as pd
from selenium.common.exceptions import NoSuchElementException, StaleElementReferenceException
from selenium.webdriver.common.action_chains import ActionChains
```

```python
driver = webdriver.Chrome()
wait = WebDriverWait(driver, 30)
```

ANDHRA PRADESH

```python
driver.get("https://www.redbus.in/online-booking/apsrtc/?utm_source=rtchometile")
```

```python
and_data = []
```

```python
def select_date(travel_day):
    try:
        # Click on the date picker to open the calendar
        date_input = driver.find_element(By.ID, 'onward_cal')
        date_input.click()

        # Wait until the calendar is visible and select the desired date
        WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.XPATH, f"//td[@class='wd day' and text()='{travel_day}']"))
        ).click()
    except Exception as e:
        print(f"An error occurred while selecting the date: {e}")
```

```python
# Step 1: Use XPath to directly select the route elements (anchor tags with the class 'route')
def scrape_page():
    route_elements = driver.find_elements(By.XPATH, '//a[@class="route"]')
    
    for route_element in route_elements:
        try:
            # Extracting the route name and link
            routename_and = route_element.text
            routelink_and = route_element.get_attribute('href')
            
            # Append data only if both route name and link are present
            if routename_and and routelink_and:
                and_data.append({
                    "Route Name": routename_and,
                    "Route Link": routelink_and
                })

                # Print for debugging purposes
                print(f"Route Name: {routename_and}")
                print(f"Route Link: {routelink_and}")
            else:
                print("Missing data for route name or route link")

        except Exception as e:
            # Handle any exceptions that may occur (like element not found or stale element)
            print(f"An error occurred: {e}")
            continue
```

```python
# Scrape data from the first 5 pages
for page_number in range(1, 6):
    scrape_page()
    if page_number < 5:  # Don't try to click next on the last page
        try:
            # Locate the pagination container
            pagination_container = wait.until(EC.presence_of_element_located(
                (By.XPATH, '//*[@id="root"]/div/div[4]/div[12]')
            ))

            # Locate the next page button within the container
            next_page_button = pagination_container.find_element(
                By.XPATH, f'.//div[contains(@class, "DC_117_pageTabs") and text()="{page_number + 1}"]'
            )

            # Ensure the next page button is in view
            actions = ActionChains(driver)
            actions.move_to_element(next_page_button).perform()
            time.sleep(1)  # Wait for a bit after scrolling

            # Log the action
            print(f"Clicking on page {page_number + 1}")

            # Click the next page button
            next_page_button.click()

            # Wait for the page number to update to the next page
            wait.until(EC.text_to_be_present_in_element(
                (By.XPATH, '//div[contains(@class, "DC_117_pageTabs DC_117_pageActive")]'), str(page_number + 1)))

            # Log the successful page navigation
            print(f"Successfully navigated to page {page_number + 1}")

            # Wait for a short duration to ensure the next page loads completely
            time.sleep(3)
        except Exception as e:
            print(f"An error occurred while navigating to page {page_number + 1}: {e}")
            break
            
```

```python
df_and = pd.DataFrame(and_data)

df_and.to_csv("/Users/jainivedhitha/Desktop/scraped_routes_and.csv", index=False)

print(df_and.head())
```

# BUS DETAILS

```python
import pandas as pd
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.chrome.service import Service  # Import Service
from selenium.webdriver.support import expected_conditions as EC

```

```python
def scrape_bus_details(state_csv, output_csv):
    # Read the CSV file containing route names and links
    df = pd.read_csv(state_csv)
    
    driver = webdriver.Chrome()  # Use Service to pass the path
    wait = WebDriverWait(driver, 60)
    
    # Lists to store bus details
    Bus_names = []
    Bus_types = []
    Departing_time = []
    Reaching_time = []
    Durations = []
    Prices = []
    Seats_Available = []
    Route_names = []
    Route_links = []
    
    # Loop through each route link in the CSV
    for index, row in df.iterrows():
        route_name = row["Route Name"]
        route_link = row["Route Link"]

        # Open the route link in the browser
        driver.get(route_link)
        time.sleep(2)
        
        # Scrape bus details using XPATH
        bus_name_elems = driver.find_elements(By.XPATH, "//div[@class='travels lh-24 f-bold d-color']")
        bus_type_elems = driver.find_elements(By.XPATH, "//div[@class='bus-type f-12 m-top-16 l-color evBus']")
        Departing_time_elems = driver.find_elements(By.XPATH, "//*[@class='dp-time f-19 d-color f-bold']")
        Reaching_time_elems = driver.find_elements(By.XPATH, "//*[@class='bp-time f-19 d-color disp-Inline']")
        duration_elems = driver.find_elements(By.XPATH, "//*[@class='dur l-color lh-24']")
        price_elems = driver.find_elements(By.XPATH, '//*[@class="fare d-block"]')
        seats_available_elems = driver.find_elements(By.XPATH, "//div[contains(@class, 'seat-left')]")
        
        # Append data to lists
        for i in range(len(bus_name_elems)):
            Bus_names.append(bus_name_elems[i].text if i < len(bus_name_elems) else "")
            Bus_types.append(bus_type_elems[i].text if i < len(bus_type_elems) else "")
            Departing_time.append(Departing_time_elems[i].text if i < len(Departing_time_elems) else "")
            Reaching_time.append(Reaching_time_elems[i].text if i < len(Reaching_time_elems) else "")
            Durations.append(duration_elems[i].text if i < len(duration_elems) else "")
            Prices.append(price_elems[i].text if i < len(price_elems) else "")
            Seats_Available.append(seats_available_elems[i].text if i < len(seats_available_elems) else "")
            Route_names.append(route_name)
            Route_links.append(route_link)
    
    # Save scraped data to a CSV file
    data = {
        'Bus_name': Bus_names,
        'Bus_type': Bus_types,
        'Start_time': Departing_time,
        'End_time': Reaching_time,
        'Duration': Durations,
        'Price': Prices,
        'Seats_Available': Seats_Available,
        'Route_name': Route_names,
        'Route_link': Route_links
    }
    df_buses = pd.DataFrame(data)
    df_buses.to_csv(output_csv, index=False)

    driver.quit()
```
